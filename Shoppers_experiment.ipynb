```python
num_exp = 4

dataset = "shoppers"
real_path = f"CHECK_DATASETS/{dataset}/trn_{dataset}_final.csv"
fake_file_root = f"Fake_Datasets/{dataset}"
categorical_columns = ['SpecialDay', 'Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend', 'Revenue']
```

```python
"""종속변수는 categorical columns와 problem_type에만 넣어야 함"""
synthesizer =  Tabular_GAN(raw_csv_path = real_path, 
                     categorical_columns = categorical_columns,
                     epochs=200, 
                     batch_size=500, 
                     bins=5,
                     private_bool = False)

for i in range(num_exp):
    synthesizer.fit()
    syn = synthesizer.generate_samples()
    syn.to_csv(fake_file_root+"/"+"Shoppers_fake_{dataset}_{exp}.csv".format(dataset = dataset, exp=i+1), index= False)
```
